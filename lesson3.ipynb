1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
from math import sqrt, factorial
selection = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

def mean(list):
    return sum(list)/len(list)

def var(list, arg):
    std = 0
    if arg==True:
        arg=1
    else:
        arg=0
    temp_mean = mean(list)
    for i in list:
        std += (i-temp_mean)**2
    return std/(len(list)-arg)


f"""Cреднее арифметическое: {mean(selection)}, 
Cреднее квадратичное отклонение: {sqrt(var(selection, False))},
Cмещенная оценка дисперсии: {var(selection, False)},
Несмещенная оценка дисперсии: {var(selection, True)}"""
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
def comb(k, n):
    return factorial(n)/(factorial(k)*factorial(n-k))

# 3 white from first
p1 = comb(3, 5)/comb(3, 8)
# 3 white from second
p2 = comb(3, 5)/comb(3, 12)
# 1 white from first and 2 from second
p3 = comb(1, 5)/comb(1, 8) * comb(2, 5)/comb(2, 12)
# 2 white from first and 1 from second
p4 = comb(2, 5)/comb(2, 8) * comb(1, 5)/comb(1, 12)
p1+p2+p3+p4
```
